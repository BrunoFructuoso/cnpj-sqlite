```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./../saidas/cnpj_5211799_uf_SP.csv', sep=',')
```

```python
df.columns
```

```python
df['cnpj'] = df['cnpj'].astype(str).str.zfill(14)
df['cep'] = df['cep'].astype(str).str.zfill(8)
df['cnae_fiscal'] = df['cnae_fiscal'].astype(str).str.zfill(7)

df['telefone_completo'] = df['ddd1'].astype('Int64').astype(str).str.zfill(2) + df['telefone1'].astype('Int64').astype(str).str.zfill(8)

df['endereco_completo'] = df.apply(
        lambda row: f"{row['tipo_logradouro']} {row['logradouro']}, {row['numero']}, {row['bairro']}, {row['municipio']} - {row['uf']}, {row['cep']}"
    if pd.notna(row['tipo_logradouro']) and pd.notna(row['logradouro']) and pd.notna(row['numero']) and pd.notna(row['bairro']) and pd.notna(row['municipio']) and pd.notna(row['uf']) and pd.notna(row['cep'])
    else None,
    axis=1
)
```

```python
df.columns
```

```python
df = df[['cnpj','nome_fantasia', 
       'data_inicio_atividades', 'cnae_fiscal', 'nome_cnae','cnae_fiscal_secundaria', 'endereco_completo',
       'tipo_logradouro', 'logradouro', 'numero', 'complemento', 'bairro',
       'cep', 'uf', 'nome_municipio', 'telefone_completo','ddd1','telefone1', 'ddd2', 'telefone2',
       ]]
```

```python
df.to_excel('./../saidas/cnpj_5211799_uf_SP_formatado.xlsx', index=False)
```
